```python
# importando bibliotecas 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dados semeados na base 42

np.random.seed(42)

#numero de registro a serem criados

n = 5000

#gerando dados

dados = pd.DataFrame({
    'id_vendas': range(1, n+1),
    'regiao': np.random.choice(['Norte','Nordeste','Centro-Oeste','Sudeste','Sul'],n),
    'categoria': np.random.choice(['Eletronicos', 'Roupas', 'Alimentos', 'Moveis','Bugigangas'],n),
    'valor_vendas': np.random.normal(500,120, n).round(2),
    'custo': np.random.normal(300,50, n).round(2),
    'satisfação_cliente': np.random.randint(1,6,n),
    'qtd_itens': np.random.randint(1,10,n)

})

dados.head()

```
