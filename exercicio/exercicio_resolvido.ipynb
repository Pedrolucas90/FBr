# Exercício Prático - Análise de Dados com Pandas e Matplotlib

## Dataset

ecommerce_5000_pedidos.csv

## Objetivo

Utilizar Pandas para explorar os dados e Matplotlib para criar visualizações que auxiliem na tomada de decisão.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('ecommerce_5000_pedidos.csv')
```

## Parte 1 - Conhecendo os Dados

1. Exiba as 5 primeiras linhas.
2. Exiba as 5 últimas linhas.
3. Verifique a quantidade de linhas e colunas.
4. Identifique os tipos de dados.

```python
print("Primeiras 5 linhas:")
display(df.head())

print("\nÚltimas 5 linhas:")
display(df.tail())

print(f"\nQuantidade de linhas: {df.shape[0]}")
print(f"Quantidade de colunas: {df.shape[1]}")

print("\nTipos de dados:")
print(df.dtypes)
```

## Parte 2 - Estatísticas Básicas

Calcule:

- Média dos pedidos
- Maior pedido
- Menor pedido
- Quantidade de produtos
- Quantidade de categorias

```python
media_pedidos = df['valor_total'].mean()
maior_pedido = df['valor_total'].max()
menor_pedido = df['valor_total'].min()
qtd_produtos = df['produto'].nunique()
qtd_categorias = df['categoria'].nunique()

print(f"Média dos pedidos: R$ {media_pedidos:.2f}")
print(f"Maior pedido: R$ {maior_pedido:.2f}")
print(f"Menor pedido: R$ {menor_pedido:.2f}")
print(f"Quantidade de produtos distintos: {qtd_produtos}")
print(f"Quantidade de categorias distintas: {qtd_categorias}")
```

## Parte 3 - Gráfico de Barras

Crie um gráfico de barras mostrando:

Quantidade de pedidos por categoria.

Personalize o gráfico com:

- Título
- Nome dos eixos
- Grade

```python
pedidos_por_categoria = df['categoria'].value_counts()

plt.figure(figsize=(8, 5))
pedidos_por_categoria.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Quantidade de Pedidos por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade de Pedidos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(pedidos_por_categoria)
```

## Parte 4 - Gráfico de Barras

Crie um gráfico de barras mostrando:

Quantidade de pedidos por cidade.

Pergunta:

Qual cidade possui mais pedidos?

```python
pedidos_por_cidade = df['cidade'].value_counts()

plt.figure(figsize=(9, 5))
plt.title('Quantidade de Pedidos por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Quantidade de Pedidos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

cidade_mais_pedidos = pedidos_por_cidade.idxmax()
print(f"A cidade com mais pedidos é: {cidade_mais_pedidos} ({pedidos_por_cidade.max()} pedidos)")
```

## Parte 5 - Gráfico de Pizza

Crie um gráfico de pizza para:

forma_pagamento

Pergunta:

Qual forma de pagamento é mais utilizada?

```python
pagamento_counts = df['forma_pagamento'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(pagamento_counts, labels=pagamento_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição das Formas de Pagamento')
plt.axis('equal')
plt.show()

forma_mais_usada = pagamento_counts.idxmax()
print(f"A forma de pagamento mais utilizada é: {forma_mais_usada} ({pagamento_counts.max()} pedidos, {pagamento_counts.max()/len(df)*100:.1f}%)")
```

## Parte 6 - Gráfico de Pizza

Crie um gráfico de pizza para:

status_pedido

Pergunta:

Qual percentual dos pedidos foi entregue?

```python
status_counts = df['status_pedido'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição do Status dos Pedidos')
plt.axis('equal')
plt.show()

pct_entregue = status_counts.get('Entregue', 0) / len(df) * 100
print(f"Percentual de pedidos entregues: {pct_entregue:.1f}%")
```

## Parte 7 - Histograma

Crie um histograma para:

valor_total

Analise:

- Concentração dos pedidos
- Distribuição dos valores

```python
plt.figure(figsize=(9, 5))
plt.hist(df['valor_total'], bins=30)
plt.title('Distribuição do Valor Total dos Pedidos')
plt.xlabel('Valor Total (R$)')
plt.ylabel('Frequência')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print(f"Valor médio: R$ {df['valor_total'].mean():.2f}")
print(f"Mediana: R$ {df['valor_total'].median():.2f}")
print(f"Desvio padrão: R$ {df['valor_total'].std():.2f}")
print("\nAnálise: a maior concentração dos pedidos está em valores mais baixos/médios, "
      "com uma cauda à direita (alguns pedidos de valor bem mais alto), indicando uma "
      "distribuição assimétrica positiva (right-skewed).")
```

## Parte 8 - Histograma

Crie um histograma para:

avaliacao_cliente

Pergunta:

Os clientes costumam avaliar bem os produtos?

```python
plt.figure(figsize=(8, 5))
plt.hist(df['avaliacao_cliente'], bins=range(1, 7))
plt.title('Distribuição das Avaliações dos Clientes')
plt.xlabel('Avaliação (1 a 5)')
plt.ylabel('Frequência')
plt.xticks([1, 2, 3, 4, 5])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

media_avaliacao = df['avaliacao_cliente'].mean()
pct_boas = (df['avaliacao_cliente'] >= 4).mean() * 100
print(f"Avaliação média: {media_avaliacao:.2f}")
print(f"Percentual de avaliações boas (4 ou 5): {pct_boas:.1f}%")
print("\nResposta: " + (
    "Sim, os clientes costumam avaliar bem os produtos." if media_avaliacao >= 3.5
    else "Não necessariamente, as avaliações estão distribuídas de forma mais equilibrada/baixa."
))
```

## Parte 9 - Boxplot

Crie um boxplot para:

valor_total

Analise:

- Outliers
- Dispersão

```python
plt.figure(figsize=(6, 6))
plt.boxplot(df['valor_total'], vert=True)
plt.title('Boxplot do Valor Total dos Pedidos')
plt.ylabel('Valor Total (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

q1 = df['valor_total'].quantile(0.25)
q3 = df['valor_total'].quantile(0.75)
iqr = q3 - q1
limite_superior = q3 + 1.5 * iqr
outliers = df[df['valor_total'] > limite_superior]
print(f"Q1: R$ {q1:.2f} | Q3: R$ {q3:.2f} | IQR: R$ {iqr:.2f}")
print(f"Limite superior para outliers: R$ {limite_superior:.2f}")
print(f"Quantidade de outliers (valores acima do limite): {len(outliers)}")
print("\nAnálise: o boxplot mostra a dispersão dos valores dos pedidos e evidencia a "
      "presença de outliers (pedidos com valores muito acima da maioria).")
```

## Parte 10 - Boxplot por Categoria

Crie um boxplot comparando:

valor_total por categoria

Pergunta:

Qual categoria possui os pedidos mais caros?

```python
plt.figure(figsize=(9, 6))
categorias = df['categoria'].unique()
dados_por_categoria = [df[df['categoria'] == cat]['valor_total'] for cat in categorias]

plt.boxplot(dados_por_categoria, labels=categorias)
plt.title('Valor Total dos Pedidos por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor Total (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

mediana_por_categoria = df.groupby('categoria')['valor_total'].median().sort_values(ascending=False)
print("Mediana do valor total por categoria:")
print(mediana_por_categoria)
print(f"\nA categoria com os pedidos mais caros é: {mediana_por_categoria.idxmax()}")
```

## Parte 11 - Gráfico de Linha

Converta a coluna data_pedido para datetime.

Crie um gráfico mostrando:

Quantidade de pedidos por mês.

```python
df['data_pedido'] = pd.to_datetime(df['data_pedido'])
df['mes'] = df['data_pedido'].dt.to_period('M')

pedidos_por_mes = df.groupby('mes').size()

plt.figure(figsize=(10, 5))
pedidos_por_mes.plot(kind='line', marker='o', color='purple')
plt.title('Quantidade de Pedidos por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Pedidos')
plt.grid(linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(pedidos_por_mes)
```

## Parte 12 - Faturamento Mensal

Crie um gráfico de linha mostrando:

Faturamento mensal.

Pergunta:

Qual foi o melhor mês de vendas?

```python
faturamento_mensal = df.groupby('mes')['valor_total'].sum()

plt.figure(figsize=(10, 5))
faturamento_mensal.plot(kind='line', marker='o', color='green')
plt.title('Faturamento Mensal')
plt.xlabel('Mês')
plt.ylabel('Faturamento (R$)')
plt.grid(linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

melhor_mes = faturamento_mensal.idxmax()
print(faturamento_mensal)
print(f"\nO melhor mês de vendas foi: {melhor_mes} (R$ {faturamento_mensal.max():.2f})")
```

## Parte 13 - Scatter Plot

Crie um gráfico de dispersão entre:

desconto x valor_total

Pergunta:

Existe relação entre desconto e valor da compra?

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['desconto'], df['valor_total'], alpha=0.4, color='crimson')
plt.title('Relação entre Desconto e Valor Total do Pedido')
plt.xlabel('Desconto (%)')
plt.ylabel('Valor Total (R$)')
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

correlacao = df['desconto'].corr(df['valor_total'])
print(f"Correlação entre desconto e valor_total: {correlacao:.3f}")
print("\nAnálise: " + (
    "Não há uma relação linear forte entre o desconto aplicado e o valor total da compra "
    "(correlação próxima de zero), ou seja, o tamanho do desconto não determina o valor final do pedido."
    if abs(correlacao) < 0.3 else
    "Existe uma relação perceptível entre o desconto e o valor total do pedido."
))
```
