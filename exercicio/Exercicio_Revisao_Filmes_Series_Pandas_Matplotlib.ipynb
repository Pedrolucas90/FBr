# Exercício de Revisão - Análise de Filmes e Séries com Pandas e Matplotlib

## Dataset

catalogo_filmes_series_5000.csv

## Objetivo

Utilizar Pandas para explorar um catálogo de filmes e séries e criar gráficos utilizando Matplotlib.

O foco é revisar análise exploratória de dados e escolha adequada de visualizações.


```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('catalogo_filmes_series_5000.csv')
```

## Parte 1 - Conhecendo os Dados

1. Exiba as 5 primeiras linhas.
2. Exiba as 5 últimas linhas.
3. Verifique quantidade de linhas e colunas.
4. Identifique os tipos de dados.
5. Verifique se existem valores nulos.


```python
print("\nPrimeiras 5 linhas:")
display(df.head())

print("\nÚltimas 5 linhas:")
display(df.tail())

print(f"\nQuantidade de linhas: {df.shape[0]}")
print(f"Quantidade de colunas: {df.shape[1]}")

print("\nTipos de dados:")
print(df.dtypes)

print("\nValores nulos:")
print(df.isnull().sum())
```

## Parte 2 - Estatísticas Básicas

Calcule:

- Média da avaliação do público.
- Maior e menor avaliação.
- Média de duração dos conteúdos.
- Quantidade de filmes.
- Quantidade de séries.


```python
media_avaliacao = df['avaliacao_publico'].mean()
maior_avaliacao = df['avaliacao_publico'].max()
menor_avaliacao = df['avaliacao_publico'].min()
media_duracao = df['duracao_minutos'].mean()
qtd_filmes = df['tipo'].nunique()
qtd_series = df['tipo'].nunique()

print(f"Média das avaliações do público: R$ {media_avaliacao:.2f}")
print(f"Maior avaliação do público: R$ {media_avaliacao:.2f}")
print(f"Menor avaliação do público: R$ {media_avaliacao:.2f}")
print(f"Média da duração dos conteúdos : R$ {media_duracao:.2f}")
print(f"Quantidade de filmes : {qtd_filmes}")
print(f"Quantidade de séries: {qtd_series}")
```

## Parte 3 - Gráfico de Barras

Crie um gráfico de barras mostrando:

Quantidade de conteúdos por tipo.

Pergunta:

Qual tipo possui maior quantidade: filmes ou séries?


```python
filmes_por_tipo = df['tipo'].value_counts()

plt.figure(figsize=(8, 5))
filmes_por_tipo.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Filmes e Série')
plt.xlabel('tipo')
plt.ylabel('Quantidade')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(filmes_por_tipo)
```

## Parte 4 - Gráfico de Barras por Gênero

Crie um gráfico mostrando:

Quantidade de conteúdos por gênero.

Analise:

Qual gênero aparece com maior frequência?


```python
qdt_por_genero = df['genero'].value_counts()

plt.figure(figsize=(9, 5))
qdt_por_genero.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Quantidade de generos')
plt.xlabel('tipos')
plt.ylabel('Quantidade')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

qdt_mais_frequente = qdt_por_genero.idxmax()
print(f"O genero com mais frequente é: {qdt_mais_frequente} ({qdt_por_genero.max()} )")
```

## Parte 5 - Gráfico de Pizza

Crie um gráfico de pizza mostrando:

Distribuição dos conteúdos por plataforma.

Pergunta:

Qual plataforma possui maior quantidade de títulos?


```python
plataforma_counts = df['plataforma'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(plataforma_counts, labels=plataforma_counts.index, autopct='%1.1f%%')
plt.title('Plataformas')
plt.axis('equal')
plt.show()

qdt_plataforma_frequente = plataforma_counts.idxmax()
print(f"A plataforma com mais frequente é: {qdt_plataforma_frequente} ({plataforma_counts.max()} )")
```

## Parte 9 - Comparação entre Filmes e Séries

Crie um boxplot comparando:

avaliacao_publico por tipo de conteúdo.

Pergunta:

Filmes ou séries possuem melhor avaliação média?


```python
plt.figure(figsize=(9, 6))
avaliacao_publico = df['avaliacao_publico'].unique()
tipo_de_conteúdo = df['tipo'] 

plt.boxplot(avaliacao_publico, labels=tipo_de_conteúdo)
plt.title('Valor Total dos Pedidos por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Valor Total (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

mediana_por_categoria = df.groupby('avaliacao_publico')['valor_total'].median().sort_values(ascending=False)
print("Mediana do valor total por categoria:")
print(mediana_por_categoria)
print(f"\nA categoria com os pedidos mais caros é: {mediana_por_categoria.idxmax()}")
```
